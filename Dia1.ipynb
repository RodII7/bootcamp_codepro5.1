```python
cantidad_caramelos = int(input('ingrese la cantidad de caramelos'))
cantidad_niños = int(input('ingrese la cantidad de niños'))
caramelos_por_niños = cantidad_caramelos // cantidad_niños
print('caramelos por niños:', caramelos_por_niños)

restantes = cantidad_caramelos % cantidad_niños
print('y sobran', restantes, 'caramelos.')
```

```python

#definir una funcion
def receta():
    nombre_receta = input('ingrese el nombre de la receta:')
    ingredientes = input('ingrese los ingredientes:')
    preparacion = input('ingrese la preparacion:')
    print('la receta es:', nombre_receta)
    print('los ingredientes son:', ingredientes)
    print('los pasos para la preparacion son:', preparacion)

receta()
receta()
receta()
```
