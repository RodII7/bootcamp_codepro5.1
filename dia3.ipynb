```python
def imprimir_palabra():
    palabra = input('ingrese la palabra que desea imprimir: ')
    print(f'la palabra ingresada es: {palabra}')

imprimir_palabra()
```

```python
letra_ingresada1 = input('ingrese la primera letra: ')
letra_ingresada2 = input('ingrese la siguinete letra: ')

def letras_ingresadas(letra_1, letra_2):
    if letra_1 == letra_2:
        print('las letras ingresadas son iguales')
    else:
        print('las letras ingresadas son diferentes')

letras_ingresadas(letra_ingresada1, letra_ingresada2)
```

```python
palabra_usuario = input('ingrese su palabra')
numero_usuario = int(input('ingrese su numero'))

def verificar(palabra_ingresada, numero_ingresado):
    if len(palabra_ingresada) == numero_ingresado:
        print('la cantidad de letras en la palabra coincide con el numero')
        return True
    else:
        print('las letras en la palabra no coinciden con el numero')
        return False

imprimir_resultado = verificar(palabra_usuario, numero_usuario)
```
